#### 12.What's the difference between a model parameter and a hyperparameter in a learning algorithm?

Ans. Model parameters are parameters of the model itself, for example slope of a linear regression model. These determines how well the model is to make predictions.

Hyperparameters are paramets of the algorithm which regularizes the learning algorithm.

 The most popular method they use to achieve success is minimizing the cost function and putting a penalty for model complexity if the model is regularized.
 
 To make predictions we feed a new instance to the parameter values of trained model using found by the learning algorithm.

#### 15.What happens if the model performs well on the training data but fails to generalize the results to new situations? Can you think of three different options?

Ans. This is a possible case of overfitting. Three possible options are: getting more data, using a simpler model to reduce the features and complexity, and reducing the noise in the data.

#### 16.What exactly is a test set, and why would you need one?

Ans. Test set is used to calculate the generalized error to new instances before using the model for production.

#### 17.What is a validation set's purpose?

Ans. Validation set is used to compare different models and find the best model and tune the hyperparameters.

#### 18.What precisely is the train-dev kit, when will you need it, how do you put it to use?

Ans. Train-dev kit is used when there is a possibility of mismatch in the training data and test, validation data. Train-dev is a part of training data that is held out and used later alongwith validation data to check the error. If the model predicts well on train-dev batch but not so well on validation data, then there is a data mismatch. If the model is performing well only on traing data but not so well on both train-dev and validatoin data, then the model is probably overfitting.

#### 19.What could go wrong if you use the test set to tune hyperparameters?

Ans. Tuning hyperparameter on test set will increase the risk of overfitting on test data, hence we'll get optimisitc model on test data but it will perform worse on real world problem.
