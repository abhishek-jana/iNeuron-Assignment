### 1. To what does a relative path refer?

Ans: Relative path refers to a location relative to the current directory.

### 2. What does an absolute path start with your operating system?

Ans. Absolute path starts with the root directory and requires the complete path of the directories and subdirectories needed to locate a file or folder.

### 3. What do the functions os.getcwd() and os.chdir() do?

Ans: os.getcwd() -- means get current working directory. Returns the absolute path of the current directory.

os.chdir() -- means change directory

### 4. What are the . and .. folders?

Ans: '.' is the current directory and '..' is the parent directory. This is useful as we don't have to type the entire path to access some file/folder.

### 5. In C:\bacon\eggs\spam.txt, which part is the dir name, and which part is the base name?

Ans: 'C:\bacon\eggs\' is the directory name and 'spam.txt' is the base name.

### 6. What are the three “mode” arguments that can be passed to the open() function?

Ans: The three modes are:

read -- used to read a file

write -- used to write in a file

append -- used to append in a file

### 7. What happens if an existing file is opened in write mode?

Ans: In write mode the existing contents are erased and a completely empty file is opened.

### 8. How do you tell the difference between read() and readlines()?

Ans: In read() the entire content of a file is read a string and in readlines() a file is read line by line. For example if we have a file test.txt with contents:

test line one

test line two

testline three

read() method will return:

test line one

test line two

testline three

and readlines() method will return:

['test line one\n', 'test line two\n', 'test line three\n']

### 9. What data structure does a shelf value resemble?

Ans: shelf value resembles a dictionary data structure. With key() and values() method work similar to dictionary methods.
